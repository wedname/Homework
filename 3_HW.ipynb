```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
np.random.seed(42)
```

```python
df = pd.read_csv("Datasets/imports-85.data.csv")
df
```

```python
df.info()
```

```python
# Удаление категориальных признаков

df_num = df.drop(['make', 'fuel-type', 'aspiration',
                'num-of-doors', 'body-style',
                'drive-wheels', 'engine-location',
                ' engine-type', ' num-of-cylinders',
                ' fuel-system', ' bore', ' stroke', 'Unnamed: 26'], axis=1)

# Подстановка наиболее частых значений в отсутствующие места с помощью класса SimpleImputer (медианные не работают)

imp = SimpleImputer(missing_values='?', strategy="most_frequent")
X = imp.fit_transform(df_num)
df_num = pd.DataFrame(X, columns=df_num.columns)
df_num
```

```python
# Удаление численных признаков

df_cat = df.drop(['symboling', 'normalized-losses', ' wheel-base', ' length', 
                    ' width', ' height', ' curb-weight', ' engine-size', ' compression-ratio', 
                    ' horsepower', ' peak-rpm', ' city-mpg', ' highway-mpg', 'Unnamed: 26'], axis=1)

# Подстановка наиболее частых значений в отсутствующие места с помощью класса SimpleImputer

imp = SimpleImputer(missing_values="?", strategy="most_frequent")
imp.fit(df_cat)
X = imp.transform(df_cat)
df_cat = pd.DataFrame(X, columns=df_cat.columns)
df_cat
```

```python
# Запись этих данных в файл

df = pd.concat([df_num, df_cat], axis=1)

df.to_csv('Datasets/cars_imputed.csv.csv')
df
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.35, random_state=42)
train_set.head(5)

```

```python
Xs = ['normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels',
      'engine-location', ' wheel-base', ' length', ' width', ' height', ' curb-weight', ' engine-type',
      ' num-of-cylinders', ' engine-size', ' fuel-system', ' bore', ' stroke', ' compression-ratio',
      ' horsepower', ' peak-rpm', ' city-mpg', ' highway-mpg', '  price']
Ys = ['symboling']
```

```python
train_Xs = train_set[Xs]
train_Ys = train_set[Ys]

test_Xs = test_set[Xs]
test_Ys = test_set[Ys]
```

```python
def MAE(Ys_real, Ys_predict):
    Losses = np.abs(Ys_real - Ys_predict)
    m = len(Ys_real)
    cost_result = 1 / m * np.sum(Losses)
    return cost_result

```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_Xs, train_Ys)
predictions = model.predict(train_Xs)
print("MAE train:", MAE(train_Ys, predictions))

```

```python
predictions = model.predict(test_Xs)
print("MAE test:", MAE(test_Ys, predictions))
```

```python
# Используя подход факторизации категориальных признаков мы увеличили точность модели (уменьшив показатели MAE train и MAE test).
```
