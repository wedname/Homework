```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "First": np.random.randint(10, size=30), 
    "Second": np.random.randint(10, 20, size=30), 
    "Third": np.random.randint(20, 30, size=30)
})

print(df)
```

```python
df.describe()
```

```python
df['First'].loc[((df['First'] % 2) != 0)] = df['First'].median()
print(df)
```

```python
df['Second'].loc[(df['Second'] % 2 == 0)] = df['Second'].mean()
print(df)
```
